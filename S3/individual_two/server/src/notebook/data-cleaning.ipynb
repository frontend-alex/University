## Validating the data 


```python
for col in df.columns:
    null_count = df[col].isnull().sum()
    if null_count > 0:
        print(f"{col}: {null_count} null values")
```

```python
(df["Age"] < 0).any()
```

```python
if not df["Intensity"].between(1, 5).all():
    count = int((~df["Intensity"].between(1, 5)).sum())
    print(count)

if not df["Location"].between(0, 4).all():
    count = int((~df["Location"].between(0, 4)).sum())
    raise ValueError(f"Invalid Location values (must be 1..4): {count}")
```
